# one_hot encoded

```python
X_encoded = pd.get_dummies(X,columns= ['model','transmission','fuelType'], drop_first = True)
```

```python
X_encoded.columns
```

```python
X_encoded = X_encoded.astype(int)
```

```python
X_encoded
```

# Label_encoded

```python
from sklearn.preprocessing import LabelEncoder

columns = ['model', 'transmission', 'fuelType']

Xlable = X.copy()  # make a safe copy
label_encoders = {}

for col in columns:
    le = LabelEncoder()
    Xlable[col] = le.fit_transform(Xlable[col].astype(str))  # Convert to string in case of nulls
    label_encoders[col] = le
```

```python
Xlable
```

```python
Xlable['model'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_col = ['year','mileage','tax','mpg','engineSize']
scaler = StandardScaler()
X_encoded[numerical_col] = scaler.fit_transform(X_encoded[numerical_col])
```

```python
X_encoded
```

```python
Xlable.columns
```

```python
Xlable[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(Xlable[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
Xlable
```

# Model train_test

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

# One_hot encoding

```python
 X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.20, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
# adjusted r2 check
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1- ((1-r2)*(n-1)/ (n-p-1))
print("Adjusted R2 score:", adjusted_r2)
```
